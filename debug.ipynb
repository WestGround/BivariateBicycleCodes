```python
import pickle
import scipy
import numpy as np
```

```python
n, k = 144, 12
error_rate = 0.001
num_cycles = 12
```

```python
orig_title = './TMP/mydata_' + str(n) + '_' + str(k) + '_p_' + str(error_rate) + '_cycles_' + str(num_cycles)
my_title = './TMP/mydata_' + str(n) + '_' + str(k) + '_p_' + str(error_rate) + '_cycles_' + str(num_cycles)

with open(orig_title, 'rb') as fp:
	orig_data = pickle.load(fp)
with open(my_title, 'rb') as fp:
	my_data = pickle.load(fp)
```

```python
for k, v1 in orig_data.items():
    v2 = my_data[k]
    print("Type of %s in original: %s"%(k, type(v1)))
    if isinstance(v1, scipy.sparse._csc.csc_matrix):
        v1_np = v1.toarray()
        v2_np = v2.toarray()
        assert np.array_equal(v1_np, v2_np)
        print("[%s] equality passed"%(k))
    elif isinstance(v1, np.ndarray):
        assert np.array_equal(v1, v2)
        print("[%s] equality passed"%(k))
    elif isinstance(v1, dict) or isinstance(v1, list):
        assert v1==v2
        print("[%s] equality passed"%(k))
    elif isinstance(v1, int) or isinstance(v1, float):
        assert v1==v2
        print("[%s] equality passed"%(k))
```
